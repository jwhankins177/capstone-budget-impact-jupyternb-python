lib 2016 - 2020

```python
lib_16_20 = dd[['FY2016 Actual Expenses','FY2017 Actual Expenses','FY2018 Actual Expenses','FY2019 Actual Expenses','FY2020 Actual Expenses']]
```

```python
lib_16_20 = lib_16_20.T
```

```python
lib_16_20['year'] = [2016,2017,2018,2019,2020]
```

```python
lib_16_20 = lib_16_20.reset_index()
```

```python
lib_16_20
```

```python
lib_16_20['year'] = lib_16_20['year'].astype('int')
```

```python
lib_16_20.plot(kind='line', x = 'year', y = 'Public Library')
```

```python
x = lib_16_20['year']
y = lib_16_20['Public Library']
```

Presentation graduation graph

```python
lib_16_20.plot(kind='line', x = 'year', y = 'Public Library')
# Calculate the Trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(lib_16_20['year'], p(x), color="purple", linewidth=3, linestyle="--")
plt.title('Library Budget by Year')
plt.xlabel('Year')
plt.ylabel('Budget')
plt.xticks(x)
```
